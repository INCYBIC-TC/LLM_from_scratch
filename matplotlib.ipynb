# Matplotlib Essential

## 1.Barplot

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=['A','B','C']
y=[80,24,57]

plt.xlabel('student',fontsize=15)
plt.ylabel('marks')
plt.title('barplot')

c=["y","b","m"]

# bar function
plt.bar(
    x,y,
    width=0.4,color=c,align='center',
    edgecolor='b',linewidth=3,linestyle=':',alpha=0.5
)
plt.show()
```

```python
x=['A','B','C','D']
y=[80,24,57,33]

plt.xlabel('student',fontsize=15)
plt.ylabel('marks')
plt.title("demo")

c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color= c,align='center',edgecolor='b',linewidth=3,linestyle =':',alpha=1.0,label="ML")
plt.legend()
plt.show()
```

```python
# double bar graph 
x=['A','B','C']
y=[80,50,77]
z=[30,34,67]

plt.xlabel('student',fontsize=15)
plt.ylabel('marks')
plt.title("demo")


plt.bar(x,y, width=0.4,color='b',label="ML")
plt.bar(x,z,width=0.4,color='m',label="DS")
plt.legend()

plt.show()
```

```python
x=['A','B','C']
y=[80,50,77]
z=[30,34,67]

width =0.2
p=np.arange(len(x))
p1 =[j+width for j in p]


plt.xlabel('student',fontsize=15)
plt.ylabel('marks')
plt.title("demo")

plt.bar(p,y,color='b',label="ML",width =0.2)
plt.bar(p1,z,color='m',label="DS",width =0.2)



plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
# horizontal bar graph
A=['A','B','C']
B=[80,50,77]
C=[30,34,67]

plt.xlabel('student',fontsize=15)
plt.ylabel('marks')
plt.title("demo")

plt.barh(A,B,color='b',label="ML",height=0.4)
plt.barh(A,C,color='m',label="DS",height=0.4)

plt.legend()
plt.show()
```

## 4.Pie Chart

```python
x=[10,20,30,40]
y=['python','java','c++','c']


plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['python','java','c++','c']

e =[0.5,0.0,0.0,0.0]
c=['r','b','y','g']


plt.pie(x,labels=y,explode=e,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['python','java','c++','c']

e =[1,0.0,0.0,0.0]
c=['r','b','y','g']


plt.pie(x,labels=y,explode=e,colors=c,autopct='%0.1f%%')
plt.show()
```

```python
x=[10,20,30,40]
y=['python','java','c++','c']

e =[0.5,0.0,0.0,0.0]
c=['r','b','y','g']


plt.pie(x,labels=y,
        explode=e,colors=c,
        autopct='%0.1f%%',shadow=True,
        radius=1.2,labeldistance=1.2,
        startangle=90,textprops={"fontsize":15},
        counterclock=False,
        wedgeprops={"linewidth":8,"edgecolor":"m"},
        center=(3,9),rotatelabels=True
       )

plt.title("demo")
plt.legend(loc=2)
plt.show()
```

```python
# dot pie
plt.pie([1])
plt.show()
```

```python
# double pie chart
x=[10,20,30,40]
x1=[15,95,17,31]
y=['python','java','c++','c']

plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.5)

plt.title("demo")
plt.legend()
plt.show()
```

## 7. Area vs Stack plot

```python

import matplotlib.pyplot as plt
import numpy as np
import random

z= np.random.randint(1,100,10)
print(z)
```

```python
x=[1,2,3,4]
area=[2,4,6,8]
area1=[8,16,32,64]
area2=[10,20,30,90]


l=['area','area1','area2']
c=['r','g','b']

plt.stackplot(x,area,area1,area2,labels=l,colors=c)
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4]
area=[2,4,6,8]
area1=[8,16,32,64]
area2=[10,20,30,90]


l=['area','area1','area2']
c=['r','g','b']

plt.stackplot(x,area,area1,area2,labels=l,colors=c,baseline='sym')
plt.legend(loc=2)
plt.title('python')
plt.xlabel('no.of student')
plt.ylabel('growth')
plt.grid()
plt.show()

#baseline='zero'
#baseline='wiggle'
```

## 9.fill between plot

```python
x = np.array([1,2,3,4,5,6])
y = np.array([1,2,3,4,5,6])

plt.plot(x,y)
plt.fill_between(x,y,color='m',where=(x>=2) & (x<=4))
plt.xlabel("python")
plt.ylabel("count")
plt.show()
```
